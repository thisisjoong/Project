```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

raw_data = pd.read_csv('/content/vgames2.csv')
raw_data

#Data 전처리 시작(필요없는 columns제거, 결측치 처리, 출고량 숫자 단위)

raw_data= raw_data.drop(['Unnamed: 0'], axis=1) #불필요 column 제거
raw_data.isnull().sum() # Year  271, Genre 50, Publisher 58

raw_data= raw_data.dropna(0) 

raw_data.dtypes #Year=float64, All_Sales = object

raw_data.head()

#Year의 이상한 년도 제대로 바꿔야 함
#K , M해결하고 숫자형 데이터로 바꿔야 함 raw_data= raw_data.dropna(0) 

raw_data['Year'] = raw_data['Year'].apply(pd.to_numeric, errors = 'coerce').fillna(0)

y = raw_data['Year'] 
#print(y.value_counts())

year_change= list()
for i in y:
  if 0<i<20:
    year_change.append(i+2000)
  elif 20<i<100:
  	year_change.append(i+1900)
  else:
    year_change.append(i)

year_change= pd.DataFrame({'year': year_change})
new_year = year_change
new_year.reset_index(drop=True, inplace=True)
raw_data.reset_index(drop=True, inplace=True)

raw_data = pd.concat([raw_data, new_year], axis=1)
raw_data= raw_data.drop(['Year'], axis=1)
raw_data   #year 끝

#K , M해결하고 숫자형 데이터로 바꿔야 함 raw_data= raw_data.dropna(0) 
# NA_Sales, EU_Sales, JP_Sales , Other_Sales

#raw_data['NA_Sales'].isin({'NA_Sales': ['K', 'M']})
#raw_data['NA_Sales'].astype(str).contains('K')

raw_data['NA_Sales'] = raw_data['NA_Sales'].str.replace('K', '000')
raw_data['NA_Sales'] = raw_data['NA_Sales'].str.replace('M', '')

raw_data['EU_Sales'] = raw_data['EU_Sales'].str.replace('K', '000')
raw_data['EU_Sales'] = raw_data['EU_Sales'].str.replace('M', '')

raw_data['JP_Sales'] = raw_data['JP_Sales'].str.replace('K', '000')
raw_data['JP_Sales'] = raw_data['JP_Sales'].str.replace('M', '')

raw_data['Other_Sales'] = raw_data['Other_Sales'].str.replace('K', '000')
raw_data['Other_Sales'] = raw_data['Other_Sales'].str.replace('M', '')


raw_data['NA_Sales'] = raw_data['NA_Sales'].apply(pd.to_numeric, errors = 'coerce').fillna(0)  #N
raw_data['EU_Sales'] = raw_data['EU_Sales'].apply(pd.to_numeric, errors = 'coerce').fillna(0)  #E
raw_data['JP_Sales'] = raw_data['JP_Sales'].apply(pd.to_numeric, errors = 'coerce').fillna(0)  #J
raw_data['Other_Sales'] = raw_data['Other_Sales'].apply(pd.to_numeric, errors = 'coerce').fillna(0)  #O





##value_change(raw_data['NA_Sales'])
##raw_data['NA_Sales'][10]


raw_data['NA_Sales'].dtypes
N = raw_data['NA_Sales']
NA_change= list()
for i in N:
  if i<1000:
    NA_change.append(i*1000000)
  elif i>1000:
  	NA_change.append(i)
  else:
    NA_change.append(i)

NA_change

E= raw_data['EU_Sales']
EU_change= list()
for i in E:
  if i<1000:
    EU_change.append(i*1000000)
  elif i>1000:
  	EU_change.append(i)
  else:
    EU_change.append(i)

J = raw_data['JP_Sales']
JP_change= list()
for i in J:
  if i<1000:
    JP_change.append(i*1000000)
  elif i>1000:
  	JP_change.append(i)
  else:
    JP_change.append(i)

O = raw_data['Other_Sales']
Other_change= list()
for i in O:
  if i<1000:
    Other_change.append(i*1000000)
  elif i>1000:
  	Other_change.append(i)
  else:
    Other_change.append(i)

new_data= pd.DataFrame({'NA_Sales_n': NA_change,
                       'EU_Sales_n' :EU_change,
                       'JP_Sales_n' : JP_change,
                        'Other_Sales_n' :Other_change})

new_data
df = pd.concat([raw_data, new_data], axis=1)
df= df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], axis=1)

#전처리 일단 끝
#필수 인사이트
# 1. 지역에 따라서 선호하는 게임 장르가 다를까 
# 2. 연도별 게임의 트렌드가 있을까
# 3. 출고량이 높은 게임에 대한 분석(플랫폼, 장르) 및 시각화 프로세스


#1-1 지역에 따라 선호하는 게임장르
sector_NA= df[['Genre', 'NA_Sales_n']]
sector_NA.sort_values(by=['Genre', 'NA_Sales_n'], ascending = [True, False])
grouped = sector_NA.groupby('Genre')
N=grouped.sum()['NA_Sales_n']  #Action          858750000.0
N.plot.barh();



sector_EU= df[['Genre', 'EU_Sales_n']]
sector_EU.sort_values(by=['Genre', 'EU_Sales_n'], ascending = [True, False])
grouped = sector_EU.groupby('Genre')
E=grouped.sum()['EU_Sales_n']  #Action 513640000.
E.plot.barh();

sector_JP= df[['Genre', 'JP_Sales_n']]
sector_JP.sort_values(by=['Genre', 'JP_Sales_n'], ascending = [True, False])
grouped = sector_JP.groupby('Genre')
J=grouped.sum()['JP_Sales_n']  #Role-Playing    342930000.0
J.plot.barh();

sector_Ot= df[['Genre', 'Other_Sales_n']]
sector_Ot.sort_values(by=['Genre', 'Other_Sales_n'], ascending = [True, False])
grouped = sector_Ot.groupby('Genre')
O=grouped.sum()['Other_Sales_n']  #Action          183990000.0
O.plot.barh();

#####1-2 가설검증
##### 귀무가설(H0) : NA의 출고량은 EU의 출고량과 크게 다르다., 대립가설(H1) : NA의 출고량은 EU의 출고량과 크게 다르지 않다.


#####df_numpy= df.to_numpy() 
#####np_NA = df_numpy[:, 5]
#####np_NA = np_NA.astype(np.float)
#####np_EU= df_numpy[:,6]
#####np_EU = np_EU.astype(np.float)




#####np.random.seed(1111)
#####np_NA= np.delete(np_NA, 0)
#####np_EU = np.delete(np_EU, 0)
#####print(np.mean(np_NA))
#####print(stats.ttest_ind(np_NA, np_EU)

#Ttest_indResult(statistic=15.529144069542578, pvalue=3.452031281877105e-54)
#귀무가설(H0) : NA의 출고량은 EU의 출고량과 크게 다르다(채택)






#2-1 연도별 게임의 트렌드
year_df = df.drop(['Publisher'], axis=1)
year_df= year_df.sort_values(by=['year'], ascending =False)
year_df['All_Sales']= year_df['NA_Sales_n']+year_df['EU_Sales_n']+year_df['JP_Sales_n']+year_df['Other_Sales_n']
year_df = year_df.drop(['NA_Sales_n',	'EU_Sales_n',	'JP_Sales_n',	'Other_Sales_n'], axis=1)
year_df
#groupe1= year_df.groupby('Platform')
pivot = year_df.pivot_table(values='All_Sales', index= 'year', columns='Platform', aggfunc=sum)
pivot= pivot.fillna(0)
pivot= pivot.replace('.', '').astype(int)

ax= sns.heatmap(pivot)
plt.title('Heatmap of Game Trend', fontsize=20)
plt.show() 

#3-1
df['All_Sales']= df['NA_Sales_n']+df['EU_Sales_n']+df['JP_Sales_n']+df['Other_Sales_n']
a= df.sort_values(by=['All_Sales'], axis=0, ascending = False) # Wii Sports	Wii	Sports	Nintendo 총 판매	82740000.0, 누적판매량 상위 5위 (Nintendo)

#출고량이 높은 플랫폼
sales = df[['Platform', 'All_Sales']]
sales_p= sales.sort_values(by=['Platform', 'All_Sales'], ascending = [True, False])
groupe1 = sales_p.groupby('Platform')
groupe1= groupe1.sum()['All_Sales']
groupe1.plot.bar();  #PS2

#출고량이 높은 지역
df_sector= df.drop(['All_Sales'], axis=1)
df_sector['NA']= df_sector['NA_Sales_n'].sum()
df_sector['EU'] = df_sector['EU_Sales_n'].sum()
df_sector['JP'] = df_sector['JP_Sales_n'].sum()
df_sector['Other']=df_sector['Other_Sales_n'].sum()

df_sector = df_sector.drop(['Name', 'Platform', 'Genre', 'Publisher','year','NA_Sales_n', 'EU_Sales_n', 'JP_Sales_n', 'Other_Sales_n'], axis=1)
df_sector = df_sector.drop_duplicates()
#df_sector.plot.bar(); #NA



#출고량 제일 높은 연도
df_year = df.drop(['Name', 'Platform', 'Genre', 'Publisher','NA_Sales_n', 'EU_Sales_n', 'JP_Sales_n', 'Other_Sales_n'], axis=1)
df_year_groupe = df_year.groupby('year')
df_year_sum = df_year_groupe.sum()['All_Sales']
df_year_sum.plot.bar(); #2008

#출고량 제일 높은 장르
df_genre= df.drop(['Name', 'Platform', 'year', 'Publisher','NA_Sales_n', 'EU_Sales_n', 'JP_Sales_n', 'Other_Sales_n'], axis=1)
df_genre_groupe = df_genre.groupby('Genre')
df_genre_sum = df_genre_groupe.sum()['All_Sales']
df_genre_sum.plot.bar(); #

#출고량이 제일 높은 회사
df_Publisher = df.drop(['Name', 'Platform', 'year', 'Genre','NA_Sales_n', 'EU_Sales_n', 'JP_Sales_n', 'Other_Sales_n'], axis=1)
df_Publisher_groupe = df_Publisher.groupby('Publisher')
df_Publisher_sum = df_Publisher_groupe.sum()['All_Sales']
df_Publisher_sum= df_Publisher_sum.reset_index()

df_Publisher_sum.loc[df_Publisher_sum['All_Sales'].idxmax()] #Publisher  Nintendo All_Sales    1.75958e+09


#my insight 









```
